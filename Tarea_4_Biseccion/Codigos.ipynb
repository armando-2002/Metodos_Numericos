### 1. Use el m茅todo de bisecci贸n para encontrar soluciones precisas dentro de $10^{-2}$ para $$ x^3 - 7x^2 + 14x - 6 = 0 $$ en cada intervalo. ### 

**a. [0;1]  
b. [1;3.2]  
c. [3.2;4]**

```python
import math

def f(x):
    return x**3 - 7*x**2 + 14*x - 6

def bisection(a, b, tol=1e-2, max_iter=100):
    """
    Funci贸n que implementa el m茅todo de bisecci贸n para encontrar una ra铆z de f(x) = 0 en el intervalo [a, b]
    dentro de la tolerancia 'tol' y con un m谩ximo de 'max_iter' iteraciones.
    """
    if f(a) * f(b) >= 0:
        print("La funci贸n no cambia de signo en el intervalo dado.")
        return None

    for i in range(max_iter):
        c = (a + b) / 2
        if abs(f(c)) < tol:
            print(f"Se encontr贸 una aproximaci贸n de la ra铆z en {i+1} iteraciones.")
            return c
        elif f(a) * f(c) < 0:
            b = c
        else:
            a = c

    print(f"El m茅todo no converge despu茅s de {max_iter} iteraciones.")
    return None

# Definici贸n de los intervalos y par谩metros
intervalos = [(0, 1), (1, 3.2), (3.2, 4)]
tol = 1e-2
max_iter = 100

# Iteraci贸n sobre cada intervalo
for a, b in intervalos:
    root = bisection(a, b, tol, max_iter)
    if root is not None:
        print(f"Una aproximaci贸n de la ra铆z en el intervalo [{a}, {b}] con tolerancia {tol} es: {root:.4f}")
        print(f"El valor de la funci贸n en la ra铆z aproximada es: {f(root):.4f}")
    print()  # L铆nea en blanco para separar resultados
```

**b. Use el m茅todo de bisecci贸n para encontrar una aproximaci贸n dentro de $10^-5$ para el primer valor positivo de  con  = tan .**

```python
import math

def bisection_method(tol):
    a = 1
    b = 2
    iterations = 0

    while (b - a) > tol:
        iterations += 1
        c = (a + b) / 2
        fc = math.tan(c) - c
        
        if abs(fc) <= tol:
            return c, iterations
        elif math.tan(a) * fc < 0:
            b = c
        else:
            a = c
    
    return (a + b) / 2, iterations

tolerance = 1e-5
approximation, iterations = bisection_method(tolerance)
print("Aproximaci贸n dentro de 10^(-5) para x = tan(x):", approximation)
print("N煤mero de iteraciones:", iterations)

```

###  1. Use el teorema 2.1 para encontrar una cota para el n煤mero de iteraciones necesarias para lograr una aproximaci贸n con precisi贸n de $10^-4$ para la soluci贸n de $x^3-x-1=0$ que se encuentra dentro del intervalo [1, 2]. Encuentre una aproximaci贸n para la ra铆z con este grado de precisi贸n. ### 

```python
def f(x):
    return x**3 - x - 1

def bisection_method(a, b, tol):
    n = 0
    while (b - a) / 2 > tol:
        midpoint = (a + b) / 2
        if f(midpoint) == 0:
            return midpoint, n
        elif f(a) * f(midpoint) < 0:
            b = midpoint
        else:
            a = midpoint
        n += 1
    return (a + b) / 2, n

# Par谩metros iniciales
a = 1
b = 2
tolerance = 1e-4

# Ejecuci贸n del m茅todo de bisecci贸n
root, iterations = bisection_method(a, b, tolerance)

print(f"La ra铆z aproximada es: {root}")
print(f"El n煤mero de iteraciones necesarias fue: {iterations}")
```
